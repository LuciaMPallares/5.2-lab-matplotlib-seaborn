#### Plot the correlation matrix using seaborn.

```python
# your code here
# Method 1 - matplotlib
# your code here
sns.set(style='white')     # estilo blanco  #COPIAR TODO TAL CUAL CND QUERAMOS USAR


mascara=np.triu(np.ones_like(corr, dtype=bool))   # genera una mascara para tapar valores


cmap=sns.diverging_palette(0, 10, as_cmap=True)   # paleta de colores


sns.heatmap(corr,
            mask=mascara,
            cmap=cmap,                          #pasas la tabla de corr y dnd el color es mas fuerte mas correlación hay
            vmax=1,
            center=0,
            square=True,
            linewidth=0.5,
            cbar_kws={'shrink': 0.5},
            annot=True
           );            
```
